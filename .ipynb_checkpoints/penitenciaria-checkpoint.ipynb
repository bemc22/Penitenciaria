```python
from tkinter import *
x = 128
y = 138
n = 5
```

```python
def crearCoord():
    coord = []
    top = n**2
    num = 2
    h = int(n/2)
    w = h
    coord.append([h,w])
    for k in range(n):
        if num > top:
            break
        t = k%2
        for j in range(k+1):
            if num > top:
                break            
            w += 0
            if(t==0):
                h += 1
            else:
                h += -1
            coord.append([h,w])
            num += 1            
        if num > top:
            break            
        for j in range(k+1):            
            if num > top:
                break                
            h += 0
            if t==0:
                w += -1
            else:
                w += 1            
            coord.append([h,w])
            num += 1            
        if num > top:
            break    
            
    return coord
```

```python
class MainWindow():
    
    def __init__(self, main):
       
        
        self.celda = PhotoImage(file = 'celda.png')  
        self.prisionero = PhotoImage(file = 'prisioner.png')
        self.policia = PhotoImage(file = "police.png")
        self.capturado = PhotoImage(file = "done.png")
        self.visitado = PhotoImage(file = "visited.png")        
        self.canvas = Canvas(main,width=x*n, height=y*n)
        self.canvas.grid(row=0, column=0)           
        self.pcount = 0        
        self.coord = []
        self.level = 0
        self.plevel = n**2 - 1        
        self.coord = crearCoord()                
        self.crearPrision()           
        cnt = int(n/2)
        pnt = n - 1       
        self.canvas.create_image(cnt*x, cnt*y,image = self.prisionero,anchor = NW)
        self.canvas.create_image(pnt*x, pnt*y,image = self.policia,anchor = NW)        
        self.button = Button(main, text="Run", command=self.onButton,width=20,height=2,font="helvetica 12")
        self.button.grid(row=1,column=0)
        
       
    def crearPrision(self):
            for current in self.coord:
                i = current[0]
                j = current[1]
                self.canvas.create_image(i*x, j*y,image = self.celda,anchor = NW)
        
        
    def onButton(self):
        # Para el Prisionero
        ix = self.coord[self.level][0]
        iy = self.coord[self.level][1] 
        self.canvas.create_image(ix*x, iy*y,image = self.visitado,anchor = NW)
        
        t = self.pcount%2
        if t == 0:
            self.level += 2
        else:
            self.level -= 1

        self.pcount += 1
        
        px = self.coord[self.plevel][0]
        py = self.coord[self.plevel][1] 
        self.canvas.create_image(px*x, py*y,image = self.visitado,anchor = NW)
        self.plevel -= 1
        
        if self.level == n**2:
            self.level = 0
            self.pcount = 0
        
        if self.plevel == -1:
            self.plevel = n**2 - 1  
        
        ix = self.coord[self.level][0]
        iy = self.coord[self.level][1]
        self.canvas.create_image(ix*x, iy*y,image = self.prisionero,anchor = NW)
        
        if(self.plevel == self.level):
            self.canvas.create_image(ix*x, iy*y,image = self.capturado,anchor = NW)  
        else:
            px = self.coord[self.plevel][0]
            py = self.coord[self.plevel][1] 
            self.canvas.create_image(px*x, py*y,image = self.policia,anchor = NW)
            
    
```

```python
root = Tk()
root.title("Penitenciaria")
MainWindow(root)
root.mainloop()
```
